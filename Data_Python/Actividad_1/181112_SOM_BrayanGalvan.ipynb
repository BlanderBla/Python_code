```python
# LIBRERIAS
import pandas as ps
import numpy as np
import matplotlib.pyplot as pl
from sklearn.preprocessing import StandardScaler
from sklearn_som.som import SOM
```

```python
# DATOS
datos = ps.read_csv("Escuelas_SOM.csv")
datos
```

```python
# DATOS A USAR
datos2 = datos[['NLLC:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro I','NLLC:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro II','NLLC:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro III','NLLC:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro IV','NLM:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro I','NLM:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro II','NLM:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro III','NLM:Porcentaje de alumnos de escuelas parecidas en cada nivel de logro IV']]
datos2
```

```python
# ARREGLO
x = np.array(datos2)
x
```

```python
# DEFINIR SOM
# GRUPOS = 4
# ENTRADA DE DATOS = 8
datos_som = SOM(m = 4, n= 1, dim = 8)
```

```python
datos_som.fit(x)
```

```python
# PRONOSTICO
pronostico = datos_som.predict(x)
pronostico
```

```python

datosF = datos2.assign(grupos = pronostico)
datosF
```

```python
Grupos = datosF.sort_values('grupos')
Grupos
```

```python
data = ps.DataFrame(Grupos)
data.to_csv('Escuelas_Kohonen.csv', index=False)
```
