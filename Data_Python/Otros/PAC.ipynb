```python
import pandas as pd
import pandas_datareader as web #Nos servira para comparar lo que se encuentra en la pantalla con lo requerido
from datetime import datetime
```

```python
fuente = "yahoo"
fecha_I= "2014-09-16"
fecha_F= datetime.today().strftime('%Y-%m-%d')
criptomoneda = input("Ingrese la criptomoneda a evaluar: ")
```

```python
df = web.DataReader(criptomoneda, data_source=fuente, start=fecha_I, end=fecha_F)
df
```
