```python
# Kmeans es un algoritmo no supervisado, por lo que no utiliza un conjunto de entrenamiento ni de prueba (algo con que compararlo)
import pandas as ps
import numpy as np
import seaborn as sb
import matplotlib.pyplot as pl
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
datos = ps.read_csv("Universidades.csv")
datos
```

```python
x = datos[['sat','top10','Aceptados','REP','Costo','grad']]
x
```

```python
kmedias = KMeans(n_clusters=3, random_state=42) # n_clusters-Numero de grupos, random_state-Semilla
```

```python
estandariza = StandardScaler()
```

```python
x = estandariza.fit_transform(x)
x
#Valores estandarizados
```

```python
kmedias.fit(x) # Ajuste
```

```python
pronostico = kmedias.predict(x)
pronostico
```

```python
datos = datos.assign(grupo = pronostico)
datos
```

```python
print(datos.sort_values(by='grupo'))
```
