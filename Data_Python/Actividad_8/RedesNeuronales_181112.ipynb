```python
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.model_selection import train_test_split
```

```python
estandariza = StandardScaler()
```

```python
datos = ps.read_csv("housing.csv")
datos.head(2)
```

```python
correlacion = datos.corr(method='pearson')
correlacion
```

```python
X_train, X_test = train_test_split(datos,test_size=0.3, random_state=42)
```

```python
entrena = X_train
prueba = X_test
```

```python
y_train = ps.DataFrame(X_train['MEDV'])
y_test = ps.DataFrame(X_test['MEDV'])
```

```python
X_train = X_train[['CRIM','ZN','INDUS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
X_test = X_test[['CRIM','ZN','INDUS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
```

```python
X_train = estandariza.fit_transform(X_train)
y_train = estandariza.fit_transform(y_train)
X_test = estandariza.fit_transform(X_test)
y_test=estandariza.fit_transform(y_test)
```

```python
tf.keras.backend.clear_session()
model=tf.keras.Sequential()
```

```python
# Utilizamos la funcion de activación relu 
model.add(Dense(5, input_dim=12, activation='relu'))     # Relu    -> (Neuronas,dim,Método de activacion)
model.add(Dense(1, activation='sigmoid'))                # Sigmoid -> (Neuronas,Método de activacion)
# Para este analisis no evaluamos con el metodo tanh ya que no nos lo pide
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
# Con el modelo obtenido, calculamos el valor de MEDV con el conjunto de prueba
model.fit(X_test, y_test, epochs=1000)
```

```python
model.summary()
```

```python
salida_train = estandariza.inverse_transform(model.predict(X_test))
```

```python
prueba = prueba.assign(pronostico = salida_train)
prueba
```

```python
prueba = prueba.assign(A = prueba['MEDV']-prueba['pronostico']) # Resta MESV y pronóstico
prueba
```

```python
prueba['A']= prueba['A'].abs() # Valor absoluto de la columna A
prueba = prueba.assign(B_2 = (prueba['A']*100)/prueba['MEDV']) # Se calcula el porcentaje de error
prueba
```
