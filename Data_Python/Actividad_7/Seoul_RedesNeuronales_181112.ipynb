```python
import sys
import tensorflow
import pandas as ps
import numpy as np
import seaborn as sb
import statistics
import matplotlib.pyplot as pl
import tensorflow as tf
```

```python
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.model_selection import train_test_split
```

```python
estandariza = StandardScaler()
```

```python
datos = ps.read_csv("SeoulBikeData.csv")
datos.head(2)
```

```python
correlacion = datos.corr(method='pearson')
correlacion
```

```python
X_train, X_test = train_test_split(datos,test_size=0.3, random_state=42)
entrena = X_train
prueba = X_test
prueba2 = X_test
prueba3 = X_test
```

```python
y_train = ps.DataFrame(X_train['RentedBike'])
y_test = ps.DataFrame(X_test['RentedBike'])
```

```python
X_train = X_train[['Hour','Temperature(C)','Dewpointtemperature(C)']]
X_test = X_test[['Hour','Temperature(C)','Dewpointtemperature(C)']]
```

```python
X_train = estandariza.fit_transform(X_train)
y_train = estandariza.fit_transform(y_train)
X_test = estandariza.fit_transform(X_test)
y_test=estandariza.fit_transform(y_test)
```

```python
tf.keras.backend.clear_session()
model=tf.keras.Sequential()
```

```python
# Entrenar 3 modelos de redes neuronales con diferente número de neuronas en capa oculta
# Primer modelo de red neuronal
model.add(Dense(5, input_dim=3, activation='relu'))# 5 neuronas en la siguiente capa ; 3 entradas (neuronas)
model.add(Dense(1, activation = 'tanh')) # 1 neurona de salida
model.compile(loss='mean_squared_error',optimizer='adam')
model.fit(X_test, y_test, epochs=1000)
```

```python
model.summary()
```

```python
salida = estandariza.inverse_transform(model.predict(X_test))
```

```python
prueba = prueba.assign(pronostico = salida)
prueba
```

```python
prueba = prueba.assign(A = prueba['RentedBike']-prueba['pronostico']) # Resta RentedBike y pronóstico
prueba['A']= prueba['A'].abs() # Valor absoluto de la columna A
prueba = prueba.assign(B_2 = (prueba['A']*100)/prueba['RentedBike']) # Se calcula el porcentaje de error
C=prueba['B_2'].mean() # Calculamos el porcentaje de error total
C
```

```python
# Segundo modelo de red neuronal
model2=tf.keras.Sequential()
model2.add(Dense(25, input_dim=3, activation='relu'))# 25 neuronas en la siguiente capa ; 3 entradas (neuronas)
model2.add(Dense(1, activation = 'tanh')) # 1 neurona de salida
model2.compile(loss='mean_squared_error',optimizer='adam')
model2.fit(X_test, y_test, epochs=1000)
```
